# Preprocessing of the Raw files

```python
import re
import pandas as pd
import string
english_punctuations = string.punctuation

train_tweets = open(PATH + "data/raw" + "/" + "train" + "_sentences" + ".txt", "r").read()
val_tweets = open(PATH + "data/raw" + "/" + "val" + "_sentences" + ".txt", "r").read()

# Removing <user>, <url>, rt, numbers and hashtags
train_tweets = re.sub('(<user>|<url>|rt|[0-9]+|\B#\w*[a-zA-Z]+\w*)', '', train_tweets)
val_tweets = re.sub('(<user>|<url>|rt|[0-9]+|\B#\w*[a-zA-Z]+\w*)', '', val_tweets)

# Removing punctuation
translator = str.maketrans('', '', english_punctuations)
train_tweets = train_tweets.translate(translator)
val_tweets = val_tweets.translate(translator)

file = open(PATH + "data/minimal_preprocessing" + "/" + "train" + "_sentences" + ".txt", "w")
file.write(train_tweets)
file.close()

file = open(PATH + "data/minimal_preprocessing" + "/" + "val" + "_sentences" + ".txt", "w")
file.write(val_tweets)
file.close()
```
